Q3. Create 10 files that contains multiple lines of random strings and file size of each file must be 5 MB.

```python
import os
import random
import string

# Create 10 files, each of ~5MB
def generate_files(num_files=10, file_size_mb=5):
    size_bytes = file_size_mb * 1024 * 1024

    for i in range(1, num_files + 1):
        filename = f"random_strings_5MB_{i}.txt"
        with open(filename, "w") as f:
            total_written = 0
            while total_written < size_bytes:
                line = ''.join(random.choices(string.ascii_letters + string.digits, k=80)) + "\n"
                f.write(line)
                total_written += len(line.encode('utf-8'))
        print(f"{filename} created. Size: {os.path.getsize(filename)} bytes")

generate_files()
```

Q4. Create 5 files of size 1GB, 2GB, 3GB, 4GB and 5GB; file contains multiple lines of random strings.

```python
import os
import random
import string

def generate_random_file(filename, size_in_gb):
    size_in_bytes = size_in_gb * 1024 * 1024 * 1024
    with open(filename, "w") as f:
        total_written = 0
        while total_written < size_in_bytes:
            # Each line will be ~81 bytes (80 characters + newline)
            line = ''.join(random.choices(string.ascii_letters + string.digits, k=80)) + "\n"
            f.write(line)
            total_written += len(line.encode('utf-8'))
        print(f"✔ Created {filename} ({total_written / (1024 * 1024 * 1024):.2f} GB)")

# Generate files from 1GB to 5GB
for size in range(1, 6):
    generate_random_file(f"random_strings_{size}GB.txt", size)
```

Q5. Convert all the files of Q4 into upper case one by one

```python
import os

# Convert each large file to uppercase line-by-line (memory-efficient)
for size in range(1, 6):
    input_file = f"random_strings_{size}GB.txt"
    output_file = f"random_strings_{size}GB_UPPER.txt"

    if os.path.exists(input_file):
        with open(input_file, "r") as infile, open(output_file, "w") as outfile:
            for line in infile:
                outfile.write(line.upper())

        print(f"✅ Converted {input_file} to uppercase → {output_file}")
    else:
        print(f"❌ File not found: {input_file}")
```

Q6: Convert All Files to Uppercase Using Multithreading

```python
import os
import threading

def convert_to_uppercase(file_path):
    output_file = file_path.replace(".txt", "_UPPER.txt")
    with open(file_path, 'r') as infile, open(output_file, 'w') as outfile:
        for line in infile:
            outfile.write(line.upper())
    print(f"✅ {file_path} → {output_file}")

file_paths = [f"random_strings_5MB_{i}.txt" for i in range(1, 11)]
threads = []

for path in file_paths:
    t = threading.Thread(target=convert_to_uppercase, args=(path,))
    t.start()
    threads.append(t)

for t in threads:
    t.join()
```

Q10. Create an automated pipeline using multi-threading for:
“Automatic Download of 100 Videos from YouTube” → “Convert it to Audio”

```python
# STEP 1: Install tools (only once)
!apt install ffmpeg -y
!pip install yt-dlp moviepy

# STEP 2: Python Code
import os
from concurrent.futures import ThreadPoolExecutor
from moviepy.editor import VideoFileClip

# STEP 3: List of 100 YouTube video URLs (for demo, repeated 10x)
video_urls = [
    "https://www.youtube.com/watch?v=Gv9_4yMHFhI",
    "https://www.youtube.com/watch?v=ukzFI9rgwfU",
    "https://www.youtube.com/watch?v=aircAruvnKk",
    "https://www.youtube.com/watch?v=8zg3H04G4UY",
    "https://www.youtube.com/watch?v=IoZGSQ07e8g",
    "https://www.youtube.com/watch?v=LHBE6Q9XlzI",
    "https://www.youtube.com/watch?v=Cr6VqTRO1v0",
    "https://www.youtube.com/watch?v=7eh4d6sabA0",
    "https://www.youtube.com/watch?v=HZGCoVF3YvM",
    "https://www.youtube.com/watch?v=kjBOesZCoqc"
] * 10  # = 100 videos

# STEP 4: Function to Download a Video
def download_video(index, url):
    filename = f"ml_video_{index+1}.mp4"
    os.system(f'yt-dlp -f best -o "{filename}" "{url}"')

# STEP 5: Function to Convert Video to MP3
def convert_to_audio(index):
    video_file = f"ml_video_{index+1}.mp4"
    audio_file = f"ml_audio_{index+1}.mp3"
    if os.path.exists(video_file):
        clip = VideoFileClip(video_file)
        clip.audio.write_audiofile(audio_file)
        print(f"✅ {audio_file}")
    else:
        print(f"⚠️ File not found: {video_file}")

# STEP 6: Download videos (parallel)
with ThreadPoolExecutor(max_workers=5) as executor:
    for i, url in enumerate(video_urls):
        executor.submit(download_video, i, url)

# STEP 7: Convert to audio (parallel)
with ThreadPoolExecutor(max_workers=5) as executor:
    for i in range(len(video_urls)):
        executor.submit(convert_to_audio, i)
```

GoogleImages” → “Rescale it to 50%”.


```python
# STEP 1: Install required packages
!pip install icrawler pillow

# STEP 2: Import modules
import os
from PIL import Image
from concurrent.futures import ThreadPoolExecutor
from icrawler.builtin import GoogleImageCrawler

# STEP 3: Create folders
os.makedirs("dog_images_original", exist_ok=True)
os.makedirs("dog_images_resized", exist_ok=True)

# STEP 4: Download 500 Dog Images
google_crawler = GoogleImageCrawler(storage={"root_dir": "dog_images_original"})
google_crawler.crawl(keyword="dog", max_num=500)
```

```python
# Function to resize one image
def resize_image(image_name):
    try:
        original_path = os.path.join("dog_images_original", image_name)
        resized_path = os.path.join("dog_images_resized", image_name)

        with Image.open(original_path) as img:
            new_size = (img.width // 2, img.height // 2)
            resized_img = img.resize(new_size)
            resized_img.save(resized_path)
            print(f"✅ Resized: {image_name}")
    except Exception as e:
        print(f"❌ Error with {image_name}: {e}")

# List all images
image_files = os.listdir("dog_images_original")

# Use ThreadPoolExecutor for multithreaded resizing
with ThreadPoolExecutor(max_workers=10) as executor:
    executor.map(resize_image, image_files)
```

Q12. Create a random dataset of 100 rows and 30 columns. All the values are defined between [1,200]. Perform
the following operations:
(i) Replace all the values with NA in the dataset defined between [10, 60]. Print the count of number
rows having missing values.
(ii) Replace all the NA values with the average of the column value.
(iii) Find the Pearson correlation among all the columns and plot heat map. Also select those columns
having correlation <=0.7.
(iv) Normalize all the values in the dataset between 0 and 10.
(v) Replace all the values in the dataset with 1 if value <=0.5 else with 0.


```python
# Q12: Data Operations with Random Dataset

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# (0) Generate Dataset: 100 rows × 30 columns, values between 1 and 200
np.random.seed(42)
data = np.random.randint(1, 201, size=(100, 30))
df = pd.DataFrame(data, columns=[f"Col_{i+1}" for i in range(30)])

# (i) Replace values in range [10, 60] with NaN and count rows with missing values
df_replaced = df.mask((df >= 10) & (df <= 60), np.nan)
missing_rows_count = df_replaced.isnull().any(axis=1).sum()
print(f"(i) Number of rows with missing values: {missing_rows_count}")

# (ii) Replace all NaN values with the mean of each column
df_filled = df_replaced.fillna(df_replaced.mean())
print("\n(ii) Replaced all NaNs with column-wise mean.")

# (iii) Compute Pearson correlation matrix and plot heatmap
correlation_matrix = df_filled.corr(method='pearson')

# Save heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False)
plt.title("Pearson Correlation Heatmap")
plt.tight_layout()
plt.savefig("pearson_heatmap.png")
plt.show()

# Select columns with all correlations ≤ 0.7
low_corr_columns = correlation_matrix.columns[(correlation_matrix.abs() <= 0.7).all()]
print(f"\n(iii) Columns with correlation ≤ 0.7: {list(low_corr_columns)}")

# (iv) Normalize all values to range [0, 10] using Min-Max Scaling
df_normalized = (df_filled - df_filled.min()) / (df_filled.max() - df_filled.min()) * 10
print("\n(iv) Normalized dataset to range [0, 10]. Sample:\n", df_normalized.head())

# (v) Replace values: 1 if <= 0.5 else 0
df_binary = df_normalized.applymap(lambda x: 1 if x <= 0.5 else 0)
print("\n(v) Binary dataset (1 if <=0.5 else 0). Sample:\n", df_binary.head())
```

Q13. Create a random dataset of 500 rows and 10 columns.
Columns 1 to 4 are defined between [-10, 10];
Columns 5 to 8 are defined between [10, 20];
Columns 9 to 10 are defined between [-100, 100].
Apply following clustering algorithms; determine the optimal number of clusters and plot distance metric
graph using each algorithm.
(i) K-Mean clustering
(ii) Hierarchical clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import silhouette_score

# Step 1: Generate the dataset
np.random.seed(42)
cols_1_4 = np.random.uniform(-10, 10, size=(500, 4))
cols_5_8 = np.random.uniform(10, 20, size=(500, 4))
cols_9_10 = np.random.uniform(-100, 100, size=(500, 2))

data = np.hstack((cols_1_4, cols_5_8, cols_9_10))
df = pd.DataFrame(data, columns=[f"Col_{i+1}" for i in range(10)])

# Step 2: Standardize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Step 3: K-Means Clustering (Elbow Method)
inertia = []
cluster_range = range(1, 11)
for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(cluster_range, inertia, 'bo-')
plt.title("K-Means Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.grid(True)
plt.tight_layout()
plt.savefig("kmeans_elbow.png")
plt.show()

# Step 4: Hierarchical Clustering Dendrogram
linkage_matrix = linkage(df_scaled, method='ward')
plt.figure(figsize=(10, 6))
dendrogram(linkage_matrix, truncate_mode='lastp', p=30)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Sample Index or Cluster Size")
plt.ylabel("Distance")
plt.tight_layout()
plt.savefig("hierarchical_dendrogram.png")
plt.show()
```

Q15. Create a random dataset of 500 rows and 5 columns:
All the values are defined between [5,10].
Perform the following operations:
(i) Perform t-Test on each column.
(ii) Perform Wilcoxon Signed Rank Test on each column.
(iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_1samp, wilcoxon, ttest_ind, ranksums

# Step 1: Generate dataset
np.random.seed(42)
data = np.random.uniform(5, 10, size=(500, 5))
df = pd.DataFrame(data, columns=[f"Col_{i+1}" for i in range(5)])

# (i) One-Sample t-Test on each column (H0: mean = 7.5)
print("=== (i) One-Sample t-Test (H0: mean = 7.5) ===")
for col in df.columns:
    stat, p = ttest_1samp(df[col], popmean=7.5)
    print(f"{col}: t-statistic = {stat:.4f}, p-value = {p:.4f}")

# (ii) Wilcoxon Signed-Rank Test on each column vs 7.5
# Note: Wilcoxon requires differences, so subtract 7.5 from each value
print("\n=== (ii) Wilcoxon Signed Rank Test (H0: median = 7.5) ===")
for col in df.columns:
    try:
        stat, p = wilcoxon(df[col] - 7.5)
        print(f"{col}: W = {stat:.4f}, p-value = {p:.4f}")
    except ValueError as e:
        print(f"{col}: Skipped (likely due to zero differences)")

# (iii) Two Sample t-Test & Wilcoxon Rank-Sum Test between Col_3 & Col_4
col3 = df["Col_3"]
col4 = df["Col_4"]

# Independent Two Sample t-Test
t_stat, t_p = ttest_ind(col3, col4)
print("\n=== (iii-a) Two-Sample t-Test (Col_3 vs Col_4) ===")
print(f"t-statistic = {t_stat:.4f}, p-value = {t_p:.4f}")

# Wilcoxon Rank-Sum Test (Mann-Whitney U alternative for independent samples)
r_stat, r_p = ranksums(col3, col4)
print("\n=== (iii-b) Wilcoxon Rank-Sum Test (Col_3 vs Col_4) ===")
print(f"statistic = {r_stat:.4f}, p-value = {r_p:.4f}")
```
